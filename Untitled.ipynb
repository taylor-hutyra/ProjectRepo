## Predicting off of stats_df

First, we need to define X and Y.  While we're here, let's get a training and testing split for the supervised methods.

```python
X = stats_df.sort_index()
Y = stats_labeled_df['lane'].sort_index()
n = len(X.columns)

X_train, X_test = train_test_split(X, test_size=0.3, random_state=42)
Y_test = Y[list(X_test.index)]
Y_train = Y[list(X_train.index)]

Y_coded, target = encode_target(pd.DataFrame(Y, columns = ['lane']), 'lane')
Y_test_coded, test_target = encode_target(pd.DataFrame(Y_test, columns = ['lane']), 'lane')
Y_train_coded, train_target = encode_target(pd.DataFrame(Y_train, columns = ['lane']), 'lane')

print(Y_train_coded.sort_index().index)
print(X_train.sort_index().index)

print('X shape:', X.shape)
print("train shape:", X_train.shape)
print("test shape:", X_test.shape)
print(len(Y))
print(len(Y_coded))
print(len(Y_test_coded))
print(len(Y_train_coded))
display(X_train.head())
print(Y_train_coded)
print(len(Y_train_coded))
```

```python
print(stats_df.index, true_pos.index)
```

```python
#display(Y_coded)
target = target[0:5]
display(target)

#plt.scatter(X.iloc[:, 0], X.iloc[:, 1], c=Y_coded['Target'], s=50, cmap='rainbow');
#plt.show()
```

### Unsupervised 

We have labels, so this is kind of silly.  But, just checking the clustering.

```python
kmeans = KMeans(n_clusters=5)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)
```

```python
mat = confusion_matrix(Y_coded['Target'], y_kmeans)
sns.heatmap(mat.T, square=True, annot=True, fmt='d', cbar=False)
plt.xlabel('true label')
plt.ylabel('predicted label');
print(metrics.classification_report(Y_coded['Target'], y_kmeans))
plt.show()
```

### Supervised

```python
model = RandomForestClassifier(n_estimators=100, random_state=0)
model.fit(X_train, Y_train_coded['Target'])

y_forest = model.predict(X_test)

mat = confusion_matrix(Y_test_coded['Target'], y_forest)
sns.heatmap(mat.T, square=True, annot=True, fmt='d', cbar=False)
plt.xlabel('true label')
plt.ylabel('predicted label');
print(metrics.classification_report(Y_test_coded['Target'], y_forest))
plt.show()

display(target)
```

```python
from sklearn.neighbors import KNeighborsClassifier
neigh = KNeighborsClassifier(n_neighbors=20)
neigh.fit(X_train, Y_train_coded['Target']) 

y_knn = neigh.predict(X_test)

mat = confusion_matrix(Y_test_coded['Target'], y_knn)
sns.heatmap(mat.T, square=True, annot=True, fmt='d', cbar=False)
plt.xlabel('true label')
plt.ylabel('predicted label');
print(metrics.classification_report(Y_test_coded['Target'], y_knn))
plt.show()

display(target)
```

```python
neigh = KNeighborsClassifier(n_neighbors=20, weights = 'distance')
neigh.fit(X_train, Y_train_coded['Target']) 

y_knn = neigh.predict(X_test)

mat = confusion_matrix(Y_test_coded['Target'], y_knn)
sns.heatmap(mat.T, square=True, annot=True, fmt='d', cbar=False)
plt.xlabel('true label')
plt.ylabel('predicted label');
print(metrics.classification_report(Y_test_coded['Target'], y_knn))
plt.show()
```

### PCA Dimensionality reduction

```python
for i in range(n):
    pca = PCA(n_components=i)
    pca.fit(X)

    #display(pca.components_)
    #display(pca.explained_variance_ratio_)
    if (sum(pca.explained_variance_ratio_) >= .90) : 
        print("At least 90% of variance explined with", i, "variables") 
        break
```

```python
X_pca = pca.transform(X)
print("original shape:   ", X.shape)
print("transformed shape:", X_pca.shape)
```
